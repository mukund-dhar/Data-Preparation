K Means Clustering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load data
data = pd.read_csv('Data.csv')

# Drop any rows with missing values
data.dropna(inplace=True)

# Extract predictor variables
X = data[['Profile_Factor', 'Symmetry', 'Crossflow', 'Flow_Velocity1', 'Flow_Velocity2', 'Flow_Velocity3', 'Flow_Velocity4', 'Speed_of_Sound1', 'Speed_of_Sound2', 'Speed_of_Sound3', 'Speed_of_Sound4', 'Signal_Strength1', 'Signal_Strength2', 'Signal_Strength3', 'Signal_Strength4', 'Signal_Strength5', 'Signal_Strength6', 'Signal_Strength7', 'Signal_Strength8', 'Signal_Quality1', 'Signal_Quality2', 'Signal_Quality3', 'Signal_Quality4', 'Signal_Quality5', 'Signal_Quality6', 'Signal_Quality7', 'Signal_Quality8', 'Gain1', 'Gain2', 'Gain3', 'Gain4', 'Gain5', 'Gain6', 'Gain7', 'Gain8', 'Transit_Time1', 'Transit_Time2', 'Transit_Time3', 'Transit_Time4', 'Transit_Time5', 'Transit_Time6', 'Transit_Time7', 'Transit_Time8']]

# Perform k-means clustering
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)
labels = kmeans.labels_

# Visualize the clustering results
fig, ax = plt.subplots()
scatter = ax.scatter(X.iloc[:,0], X.iloc[:,1], c=labels, s=50)
ax.set_title('K-Means Clustering')
ax.set_xlabel('Profile_Factor')
ax.set_ylabel('Symmetry')
legend = ax.legend(*scatter.legend_elements(), loc="lower left", title="Clusters")
ax.add_artist(legend)
plt.show()
```

```python
# Perform elbow method to determine optimal number of clusters
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=0).fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Within-Cluster Sum of Squares')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering

# Load data
data = pd.read_csv('Data.csv')

# Drop any rows with missing values
data.dropna(inplace=True)

# Extract predictor variables
X = data[['Profile_Factor', 'Symmetry', 'Crossflow', 'Flow_Velocity1', 'Flow_Velocity2', 'Flow_Velocity3', 'Flow_Velocity4', 'Speed_of_Sound1', 'Speed_of_Sound2', 'Speed_of_Sound3', 'Speed_of_Sound4', 'Signal_Strength1', 'Signal_Strength2', 'Signal_Strength3', 'Signal_Strength4', 'Signal_Strength5', 'Signal_Strength6', 'Signal_Strength7', 'Signal_Strength8', 'Signal_Quality1', 'Signal_Quality2', 'Signal_Quality3', 'Signal_Quality4', 'Signal_Quality5', 'Signal_Quality6', 'Signal_Quality7', 'Signal_Quality8', 'Gain1', 'Gain2', 'Gain3', 'Gain4', 'Gain5', 'Gain6', 'Gain7', 'Gain8', 'Transit_Time1', 'Transit_Time2', 'Transit_Time3', 'Transit_Time4', 'Transit_Time5', 'Transit_Time6', 'Transit_Time7', 'Transit_Time8']]

# Perform hierarchical clustering
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.title('Hierarchical Clustering')
plt.xlabel('Predictor Variables')
plt.ylabel('Euclidean Distances')
plt.show()

# Fit agglomerative clustering model
hc = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
labels = hc.fit_predict(X)

# Visualize the clustering results
fig, ax = plt.subplots()
scatter = ax.scatter(X.iloc[:,0], X.iloc[:,1], c=labels, s=50)
ax.set_title('Hierarchical Clustering')
ax.set_xlabel('Profile_Factor')
ax.set_ylabel('Symmetry')
legend = ax.legend(*scatter.legend_elements(), loc="lower left", title="Clusters")
ax.add_artist(legend)
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.mixture import GaussianMixture

# Load data
data = pd.read_csv('Data.csv')

# Drop any rows with missing values
data.dropna(inplace=True)

# Extract predictor variables
X = data[['Profile_Factor', 'Symmetry', 'Crossflow', 'Flow_Velocity1', 'Flow_Velocity2', 'Flow_Velocity3', 'Flow_Velocity4', 'Speed_of_Sound1', 'Speed_of_Sound2', 'Speed_of_Sound3', 'Speed_of_Sound4', 'Signal_Strength1', 'Signal_Strength2', 'Signal_Strength3', 'Signal_Strength4', 'Signal_Strength5', 'Signal_Strength6', 'Signal_Strength7', 'Signal_Strength8', 'Signal_Quality1', 'Signal_Quality2', 'Signal_Quality3', 'Signal_Quality4', 'Signal_Quality5', 'Signal_Quality6', 'Signal_Quality7', 'Signal_Quality8', 'Gain1', 'Gain2', 'Gain3', 'Gain4', 'Gain5', 'Gain6', 'Gain7', 'Gain8', 'Transit_Time1', 'Transit_Time2', 'Transit_Time3', 'Transit_Time4', 'Transit_Time5', 'Transit_Time6', 'Transit_Time7', 'Transit_Time8']]


# Fit Gaussian mixture models with different numbers of components
n_components_range = range(1, 10)
models = [GaussianMixture(n, covariance_type='full', random_state=0).fit(X) for n in n_components_range]

# Calculate the Bayesian information criterion (BIC) for each model
bics = [model.bic(X) for model in models]

# Choose the model with the lowest BIC value
best_model_idx = np.argmin(bics)
best_model = models[best_model_idx]

# Print the chosen number of components and the corresponding BIC value
print(f"Chosen number of components: {n_components_range[best_model_idx]}")
print(f"BIC value for chosen model: {bics[best_model_idx]}")
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.cluster import hierarchy

# Load data
data = pd.read_csv('Data.csv')

# Drop any rows with missing values
data.dropna(inplace=True)

# Extract predictor variables
X = data[['Profile_Factor', 'Symmetry', 'Crossflow', 'Flow_Velocity1', 'Flow_Velocity2', 'Flow_Velocity3', 'Flow_Velocity4', 'Speed_of_Sound1', 'Speed_of_Sound2', 'Speed_of_Sound3', 'Speed_of_Sound4', 'Signal_Strength1', 'Signal_Strength2', 'Signal_Strength3', 'Signal_Strength4', 'Signal_Strength5', 'Signal_Strength6', 'Signal_Strength7', 'Signal_Strength8', 'Signal_Quality1', 'Signal_Quality2', 'Signal_Quality3', 'Signal_Quality4', 'Signal_Quality5', 'Signal_Quality6', 'Signal_Quality7', 'Signal_Quality8', 'Gain1', 'Gain2', 'Gain3', 'Gain4', 'Gain5', 'Gain6', 'Gain7', 'Gain8', 'Transit_Time1', 'Transit_Time2', 'Transit_Time3', 'Transit_Time4', 'Transit_Time5', 'Transit_Time6', 'Transit_Time7', 'Transit_Time8']]


# Calculate the correlation matrix
corr_matrix = X.corr()

# Perform hierarchical clustering on the correlation matrix
dendrogram = hierarchy.dendrogram(hierarchy.linkage(corr_matrix.values, method='ward'), 
                                  labels=corr_matrix.columns)

# Plot the dendrogram
plt.title('Variable Clustering - Dendrogram')
plt.ylabel('Distance')
plt.show()

# Create a heatmap of the correlation matrix
sns.heatmap(corr_matrix, cmap='coolwarm', center=0, annot=True)
plt.title('Variable Clustering - Correlation Matrix')
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.ensemble import RandomForestClassifier

# Load the data into a pandas dataframe
df = pd.read_csv('Data.csv')

# Separate the predictor variables and the target variable
X = df.drop('Health_State_of_Meter', axis=1)
y = df['Health_State_of_Meter']

# Convert the categorical target variable to numerical values
y = y.map({'Healthy': 0, 'Gas injection': 1, 'Installation effects': 2, 'Waxing': 3})

# ANOVA F-test for feature selection - variable statistics-based method
k = 5
selector = SelectKBest(score_func=f_classif, k=k)
X_selected = selector.fit_transform(X, y)

# Random Forest Classifier for feature selection - model based
model = RandomForestClassifier()
model.fit(X, y)
importance = model.feature_importances_
indices = importance.argsort()[-k:][::-1]
selected_features = X.columns[indices]

# Print the selected variables
print(f'Selected features using ANOVA F-test: {X.columns[selector.get_support()].tolist()}')
print(f'Selected features using Random Forest Classifier: {selected_features.tolist()}')
```
