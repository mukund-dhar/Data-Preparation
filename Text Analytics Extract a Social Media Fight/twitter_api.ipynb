```python
import keys
import tweepy
import pandas as pd

api_key = keys.api_key
api_secrets = keys.api_secrets
access_token = keys.access_token
access_secret = keys.access_secret
 
# Authenticate to Twitter
auth = tweepy.OAuthHandler(api_key,api_secrets)
auth.set_access_token(access_token,access_secret)
 
api = tweepy.API(auth)
 
try:
    api.verify_credentials()
    print('Successful Authentication')
except:
    print('Failed authentication')
```

```python
def get_tweets(username, num_of_tweets=500):
    """
    returns a dataframe of tweets information of the given username 
    """
    all_tweets = []
    tweets = api.user_timeline(screen_name=username, 
                           count=1,
                           include_rts = False,
                           tweet_mode = 'extended'
                           )
    all_tweets.extend(tweets)
    oldest_id = tweets[-1].id
    while len(all_tweets) < num_of_tweets:
        tweets = api.user_timeline(screen_name=username, 
                            count=200,              # 200 is the maximum allowed count
                            include_rts = False,
                            max_id = oldest_id - 1,
                            tweet_mode = 'extended' # Necessary to keep full_text otherwise only the first 140 words are extracted
                            )
        if len(tweets) == 0:
            break

        oldest_id = tweets[-1].id
        all_tweets.extend(tweets)
        print(f'N of tweets downloaded till now {len(all_tweets)}', end='\r')
    
    if len(all_tweets) >= num_of_tweets:
        all_tweets = all_tweets[:num_of_tweets]

    #transform the tweepy tweets into a 2D array that will populate the csv	

    outtweets = [[tweet.id_str, 
                tweet.created_at, 
                tweet.favorite_count, 
                tweet.retweet_count, 
                tweet.full_text.encode("utf-8").decode("utf-8")] 
                for idx,tweet in enumerate(all_tweets)]
    df = pd.DataFrame(outtweets,columns=["id","created_at","favorite_count","retweet_count", "text"])

    return df
```

```python
username1 = "Cobratate"

df1 = get_tweets(username1, num_of_tweets=3000)
df1.to_csv(f'{username1}_tweets.csv', index=False, encoding='utf-8-sig')
print(f"Total number of tweets collected for {username1} : {df1.shape[0]}")
df1.head(3)
```
